```python
from scipy.stats import zscore
```

```python
df['trestbps_zscore'] = zscore(df['trestbps'])
```

```python
anomaly_threshold = 3
df['anomaly'] = np.where(abs(df['trestbps_zscore']) > anomaly_threshold, 1, 0)
```

```python
sns.scatterplot(data=df, x='age', y='trestbps', hue='anomaly')
```
